```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Linear Relationships

```python
x = np.random.randn(100).reshape(-1,1)
y = np.random.randn(100).reshape(-1,1)
plt.scatter(x,y)
```

```python
y = .5* np.random.randn(100,1) + x
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()
regr = LR.fit(x,y)
y_pred = regr.predict(x)
plt.scatter(x,y)
plt.plot(x,y_pred,'r')
```

# Fitting random data

```python
x = np.random.randn(100).reshape(-1,1)
y = np.random.randn(100).reshape(-1,1)
plt.scatter(x,y);
LR = LinearRegression()
regr = LR.fit(x,y)
pred = regr.predict(x)
plt.plot(x,pred,'r');
f = np.var(x, ddof=1)/np.var(y, ddof=1)
print('F=', f)
```

```python
# Simulate a few thousand random datasets: 
F = []

for i in range(10000):
    x = np.random.randn(100).reshape(-1,1)
    y = np.random.randn(100).reshape(-1,1)
    f = np.var(x, ddof=1)/np.var(y, ddof=1)
    F.append(f)
```

```python
plt.hist(F,100);
```
